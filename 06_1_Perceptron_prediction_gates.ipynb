### Perceptrons - Making Predictions
#### Creating a gate with Perceptron

```python
import numpy as np #python 안에 수학적인 계산을 쓸 수 있게 하는거 numpy
```

### AND Gate

```python
def AND(x1, x2):
    x = np.array([x1, x2]) 
    w = np.array([0.5, 0.5]) #가중치는 보통 다 더해서 1이 되도록 설정한다. 
    #x = np.array([0.1, 0.9]) 
    b = -0.7 #-2 #b는 평소의 기저상태이다.
    #activation 계산
    tmp = w[0]*x[0] + w[1]*x[1] + b # tmp = np.sum(w*x) + b,  np를 이렇게 쓸 수도 있다.
    
    print(tmp) #추가함
    
    if tmp <= 0:
        return 0
    else:
        return 1
```

```python
AND(1,0)
```

```python

AND(1,1)
```

```python

AND(0,0)
```

```python
AND(0,1)
```

### NAND Gate

```python
def NAND(x1, x2):
    x = np.array([x1, x2])
    w = np.array([-0.5, -0.5])
    b = 0.7
    tmp = w[0]*x[0] + w[1]*x[1] + b #w[0]*x[0] + w[1]*x[1] + b 와 b만 다르고 똑같다.***그 이유 시험*** #아 이거 or 필기인가??ㄷㄷ
    if tmp <= 0:
        return 0
    else:
        return 1
```

```python
NAND(0,0)
```

```python
NAND(1,0)
```

```python
NAND(1,1)
```

### OR Gate

```python
def OR(x1, x2):
    x = np.array([x1, x2])
    w = np.array([0.5, 0.5])
    b = -0.2
    tmp = np.sum(w*x) + b
    if tmp <= 0:
        return 0
    else:
        return 1
```

```python
OR(0,0)
```

```python
OR(1,0)
```

```python
OR(0,1)
```

```python
OR(1,1)
```

### XOR Gate

```python
def XOR(x1, x2):
    s1 = NAND(x1, x2)
    s2 = OR(x1, x2)
    y = AND(s1, s2)
    return y
```

```python
XOR(0,0)
```

```python
XOR(1,1)
```

```python
XOR(0,1)
```

```python
XOR(1,0)
```

XOR cannot be expressed as a single layer Perceptron.
