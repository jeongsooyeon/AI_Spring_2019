```
import tensorflow as tf
mnist = tf.keras.datasets.mnist #minist를 불러온다. # keras: 인간에게 편안하게 다가온다. 즉 쉽다. 구글에서 먹어버렸다.
```

```
(x_train, y_train),(x_test, y_test) = mnist.load_data()  # 엠니스트에서 데이터를 가져온다.
x_train = x_train / 255.0 # 계산하기 편하게, 255로 나눈다.
x_test = x_test / 255.0  
```

```
x_train.shape # 항상 이렇게 확인해야한다. # test가 6000만개 들어있다.
```

```
x_test.shape # test가 만개 들어있다.
```

파라미터의 개수가 0개.
784와 10을 이어주는 거미줄이 있다.
트레이터블 파라미터 개수가 7850개.

```
model = tf.keras.models.Sequential([   #지금까지 배웠던 머신러닝 여기안에 다 들어있다. # mnist데이터는 28 * 28 이다. 텐서플로우에서는 784로 한줄로 만들어서 저장한다.
  tf.keras.layers.Flatten(input_shape=(28, 28)), # 케라스는 그렇지 않고 28 * 28 로 가지고 있는데, flatten 즉 편편하게 해서 짜부시킨다. #여기를 지나면 784가된다. 
  tf.keras.layers.Dense(512, activation=tf.nn.relu), #794개와 10개 사이에 hidden 노드를 둔다. #이쪽 #을 지우면 문제푸는 데 시간이 더 오래 걸린다. 거미줄이 많아지니까. #중간에 relu 같은 거 의미를 알면 좋겠다. # 512 = 2의 9승
  tf.keras.layers.Dense(256, activation=tf.nn.relu),
  tf.keras.layers.Dense(256, activation=tf.nn.relu),
    
  tf.keras.layers.Dense(10, activation=tf.nn.softmax) # 덴스는 거미줄이라고 생각하면된다. 784가 있고 10개를 거미줄 처럼 밀도있게 연결한다. #확률 10개에 담긴다.
])
#여기까지 거미줄 만드는 부분까지가 중요하다
```

```
# 모델을 푸는 방법을 지정해줌
model.compile(optimizer='adam',  #모델을 옵티마이저, 즉 문제를 푼다. 문제를 푸는 거에 대한 정의를 해준다.
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

아담: 옵티마이저를 똑똑하게 해주는거. 믿고 쓰는 아담.

단순한 문제는 그레디언트 디센트로도 잘 풀린다.

공 굴리기: 모멘텀 momenterm

loss : 빼거 

확률을 담는 그릇: 크로스엔트로피

미분: gradient 

```

model.fit(x_train, y_train, epochs=5) #모델을 맞춰라. #에포크가 5번 돌리라는거.

# 한 에포크의 뜻 : 있는 데이터를 다 소진하는 거. 5에포크는 따라서 5번 돌리는거.

 
#60000/60000 [==============================] - 3s 49us/sample - loss: 0.2669 - acc: 0.9261 #돌리면 이렇게 나오는데, 뒤에 0.9261이 확률이다.
#60000/60000 [==============================] - 17s 278us/sample - loss: 0.0277 - acc: 0.9917 #히든 노드를 두니까 확률이 많이 올라갔다.

#여기서 에퀴리시는 트레이닝 에퀴리시이다. 
```
