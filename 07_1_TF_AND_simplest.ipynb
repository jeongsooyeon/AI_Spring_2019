### 중요  : AND학습하기   아까 같은 문제를 텐서플로우로 풀어보자


```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
tf.set_random_seed(777)  # for reproducibility
learning_rate = 0.1
```

```python
# AND GATE 
x_data = [[0, 0],
          [0, 1],
          [1, 0],
          [1, 1]]

y_data = [[0],
          [0],
          [0],
          [1]]
'''
결과
0 [[0.00085747]]
1 [[0.08177343]]
2 [[0.08177343]]
3 [[0.90235615]]
이렇게 나온다.
'''
```

```python
# XOR    앤드게이트를 학습할 수 있는 데이터로 xor을 돌려본다.
'''
x_data = [[0, 0],
          [0, 1],
          [1, 0],
          [1, 1]]

y_data = [[0],
          [1],
          [1],
          [0]]
'''

#실행시키면 아래에서 cost가 점점 늘어난다. 또, 결과보기에서 값이 다 0.5가 나온다.and gate일땐 0에 가까운 숫자 나왔는데
```

```python
x_data = np.array(x_data, dtype=np.float32)
y_data = np.array(y_data, dtype=np.float32)
```

```python
x_data.shape
```

```python
y_data.shape
```

```python
X = tf.placeholder(tf.float32, [None, 2])
Y = tf.placeholder(tf.float32, [None, 1])

W1 = tf.Variable(tf.random_normal([2, 1]), name='weight1') #웨이트가 2개의 숫자
b1 = tf.Variable(tf.random_normal([1]), name='bias1') #바이어스가 1개의 숫자  따라서 2+1=3개의 숫자
hypothesis = tf.sigmoid(tf.matmul(X, W1) + b1)
```

```python
cost = tf.reduce_mean(tf.square(hypothesis - Y))
train = tf.train.GradientDescentOptimizer(learning_rate=learning_rate).minimize(cost)
```

```python
# Launch graph
sess = tf.Session()
```

```python
# TensorFlow 변수들(variables) 초기화 (Initialization)
sess.run(tf.global_variables_initializer())
```

```python
for i in range(10001):
    sess.run(train, feed_dict={X: x_data, Y: y_data})

    if i % 1000 ==0:
        c1 = sess.run(cost, feed_dict={X: x_data, Y: y_data})
        print('step={} / cost={}'.format(i, c1))
```
