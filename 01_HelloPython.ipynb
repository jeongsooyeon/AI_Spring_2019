## Basic data types

```python
x = 3
print(type(x)) # Prints "<class 'int'>"
print(x) # Prints "3"
print(x + 1) # Addition; prints "4"
print(x - 1) # Subtraction; prints "2"
print(x * 2) # Multiplication; prints "6"
print(x ** 2) # Exponentiation; prints "9"
x += 1
print(x) # Prints "4" # 이 때는 x값이 바뀌어서 저장
x -= 1
print(x) # Prints "3"
x *= 2
print(x) # Prints "6"
```

```python
y = 2.5
print(type(y)) # Prints "<class 'float'>"
print(y, y + 1, y * 2, y ** 2) # Prints "2.5 3.5 5.0 6.25"
```

## Operators 연산

### + operator

```python
print((1, 2, 3) + (4, 5, 6)) #더하기는 그냥 붙여주는 거
print([1, 2, 3] + [4, 5, 6])
print("Hello" + " " + "World")
```

### * operator
The * operator produces a new tuple, list, or string that "repeats" the original content.

```python
print((1, 2, 3) * 3)
print([1, 2, 3] * 3)
print("Hello "*3)
```

## Containers *중요
Python includes several built-in container types: lists, dictionaries, sets, and tuples.

### Tuple
A simple immutable (변경할 수 없는, 불변의) ordered sequence of items

```python
# creating a tuple
months = ('January','February','March','April','May','June',\
'July','August','September','October','November','December') #튜플안에 들어있다

print(months[0])
print("index of 7 ==> " , months[7])
```

To print one by one

```python
# iterate through them:
for item in months:
    print (item)
print('----')
for i in range(12):
    print(months[i])
               
```

```python
t = ('john', 32, (2,3,4,5), 'hello')
print(t)
print(t[2])
print(t[2][1])
print(t[:2]) # index NOT included **시험 자주 내시는 부분 ' : ' 
print(t[2:]) # index included O
```

```python
print(t[-1])
print(t[-2])
print(t[-3])
print(t[-4])
```

```python
t[2] = 1
```

### List
Mutable (바꿀수 있는, 변경가능한) ordered sequence of items of mixed types

```python
li = ['hallym', 1, 3.141572, 'hello']
print(li)
li[1] = 45
print(li)
li.append('September') # 뒤에 끼워넣는다
print(li)
li[0] = ('well')
print(li)
li.append('Bye')
print(li)
```
