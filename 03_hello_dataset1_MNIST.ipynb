```python
import numpy as np # numpy를 np로 가져와라. np는 np1, np2..로 만들 수 있다.  
import matplotlib.pyplot as plt # plt: 그림을 그릴때 슨다. 그림을 그리게 해 주는 것 중에 pyplot을 가져온다. 
from tensorflow.examples.tutorials.mnist import input_data # 텐서플로우에 있는 공식적인 데이터 셋을 가져왔다. 뒤에 as가 없다 즉 static으로 쓴다는 것.
```

* 다운로드받기

```python
mnist = input_data.read_data_sets("MNIST_data/", one_hot=True)
```

```python
print(mnist.train.images.shape) # mnist: 아래 그림의 데이터를 모아놓은 우체국이다. 데이터 셋, train 데이터: 공부해보는 데이터 55000 **개수를 앞에 배치한다.
print(mnist.test.images.shape) # shape: 데이터 형상을 찍기, test 데이터: 공부한 것 중에서 시험 10000 
```

28 * 28 = 784
0에서 54999까지 꺼내올 수 있다.

```python
print(mnist.train.labels.shape)
print(mnist.test.labels.shape)
```

0에서 9까지의 확률을 담을 그릇이 10개가 있다는 것이다.

텐서플로우의 확률이 그릇에 담긴다.

```python
idx = 777  #10000-1 #본인이 원하는 숫자로 바꾸어보기 #0에서 9999 까지의 자유로운 숫자 넣을 수 있다.(10000개 까지니까)
```

```python
# idx = 10000 개로 하면 에러가 난다. IndexError: index 10000 is out of bounds for axis 0 with size 10000
# idx = 9999 개는 제대로 실행됨. run : 6
```

```python
'mnist.test.images[{}]'.format(idx) # 이미지 위에 다가 글자 박고 싶어서 한거
```

```python
img1 = mnist.test.images[idx] # test를 train이라고 바꿔도 된다.
img1 = np.array(img1, dtype='float') #그림을 그리기 위해 넣었다. 이미지는 소숫점이 아닌가보다라고 추측 가능.

pixels = img1.reshape((28, 28)) # shape을 바꾼다. # 28*28 = 784 **기억해 두기, 각각의 글자는 28*28이다. 정보가 784개고 28개로 쪼갠다.
plt.imshow(pixels, cmap='gray') #plt에게 그로도록 한다. 흑백으로 그려다오
plt.title('mnist.test.images[{}]'.format(idx)) # 이미지로 저장하게 해줌
plt.show() # 그림을 그려라.
```

```python
mnist.test.labels[idx] # 1: 100%, 0: 0%, 하나만 다른것: one-hot encoding이라고 한다.
```

데이터가 쉬우면 1을 뱉어낼 수 도 있다.
근데, 헷깔릴때면 0.4 등등 으로뱉어냄. 나머지 0.6은 다른 2번째 확률로 뱉어냄.

```python
label1 = mnist.test.labels[idx] #**라벨링을 왜하느냐? 우리는 수퍼바이즈드를 하니까.
print(label1)
```

```python
np.argmax(label1) #마지막 단계는 항상 이거다. 아그멕스**. 가장 높은 확률의 값을 끄집어 내는 것.
```

```python
np.argmax([0, 1, 6, 3, 2, -2, 5, 7, 10]) # argmax 함수, 괄호로 되어있으면 함수다. 함수 안에 리스트 있다. #arg는 자리를 찾아주는거.
```

```python
idx = 38 #idx = index
img1 = mnist.test.images[idx]
img1 = np.array(img1, dtype='float')
lbl1 = np.argmax(mnist.test.labels[idx]) #0, 1 이렇게 들어잇는 확률의 그릇

pixels = img1.reshape((28, 28))
plt.imshow(pixels, cmap='gray')
# plt.title('mnist.test.images[{}] ---> {}'.format(idx, lbl1))
plt.title('Testimages[{}] ---> {}'.format(idx, lbl1))
plt.show()
```
